```python
# importing pandas library as 'pd'
import pandas as pd
```

```python
# create a simple dataframe for the example

exampledata = {
    'Name': ['Brian', 'Bob', 'Bonnie', 'Brick', 'Bhris'],
    'Age': [27, 21, 26, 37, 29],
    'City': ['Atlanta', 'Houston', 'New York', 'New Jersey', 'Orlando'],
    'Salary': [77494, 86096, 103852, 51052, 128411]
}

df = pd.DataFrame (exampledata)
```

```python
# look at example data
print(df)
```

```python
# what if the company decided to give everybody a 20% raise?
df["New_Salary"] = df["Salary"] * 1.20
```

```python
# look at new example data
print(df)
```

```python
# import matplotlib as 'plt'
import matplotlib.pyplot as plt
```

```python
# configure the size of the plot
plt.figure(figsize=(8, 4))
# Choose the column from the dataframe to plot and the color of the plt
plt.bar(df['Name'], df['New_Salary'], color='lightblue')
# title of the plot is 'Employee Salaries'
plt.title('Employee Salaries')
# x axis is the name
plt.xlabel('Name')
# y axis is the salary
plt.ylabel('Salary (£)')
```
